# TÌM HIỂU VỀ CÔNG TY, GIÁ CỔ PHIẾU VÀ DỮ LIỆU PHÂN TÍCH

## 1. Về Công ty Cổ phần FPT

Dựa trên phân tích thống kê toàn diện 8 bộ dữ liệu tài chính (Bảng Cân Đối, KQKD, và các chỉ số), bức tranh tổng quan về FPT hiện lên rất rõ nét:

- **Một "Cỗ máy Tăng trưởng Kép" (Compounder):** FPT là một doanh nghiệp tạo ra lợi nhuận cực kỳ hiệu quả và ổn định. Chỉ số `ROE` (Lợi nhuận trên Vốn chủ sở hữu) được duy trì một cách nhất quán ở mức rất cao, trung vị **~26%** qua nhiều năm.
- **Một "Pháo đài Tài chính":** Công ty có sức khỏe tài chính vượt trội. FPT hoạt động với vị thế **"tiền ròng" (net cash)**, nghĩa là lượng tiền mặt nắm giữ (Median ~75% Nợ ngắn hạn) lớn hơn toàn bộ nợ vay tài chính. Rủi ro thanh khoản gần như bằng 0, với khả năng thanh toán lãi vay (Median ~17 lần) cao gấp nhiều lần chi phí lãi vay.
- **Mô hình kinh doanh đặc thù:** Mô hình kinh doanh (đặc biệt là xuất khẩu phần mềm) đòi hỏi vốn lưu động lớn. `Vòng quay tiền mặt` rất dài (Median ~187 ngày), chủ yếu do mất nhiều thời gian thu tiền từ khách hàng (Median ~249 ngày). Để tài trợ cho điều này, FPT sử dụng chiến lược **chiếm dụng vốn** rất hiệu quả từ nhà cung cấp (Median ~179 ngày).
- **Chiến lược tăng trưởng:** FPT tăng trưởng lợi nhuận không phải bằng cách cải thiện biên lợi nhuận (vốn đã rất ổn định ở mức ~14-15%), mà bằng cách **duy trì biên lợi nhuận đó và liên tục mở rộng quy mô doanh thu**.

## 2. Về Cổ phiếu (Mã FPT)

- **Xu hướng dài hạn:** Biểu đồ giá (2021-2025) cho thấy một **xu hướng tăng giá (uptrend) rất mạnh mẽ**. Giá cổ phiếu đã tăng khoảng 4 lần, từ vùng ~38.500 VNĐ lên đỉnh ~154.000 VNĐ.
- **Vấn đề Pha loãng (Dilution):** Cần lưu ý rằng sự tăng giá này đi kèm với việc phát hành cổ phiếu liên tục (chia tách, ESOP). `Số lượng cổ phiếu lưu hành` cũng đã tăng ~3.7 lần (từ 397 triệu lên 1.471 triệu CP). Đây là lý do chính khiến `Tăng trưởng EPS` (lợi nhuận/cổ phiếu) lại chậm hơn `Tăng trưởng Lợi nhuận sau thuế` của toàn công ty.
- **Các Yếu tố Ảnh hưởng (Risk Factors) (Từ Phân tích Tương quan):**
  - Giá cổ phiếu FPT **tương quan thuận** mạnh với các chỉ số cơ bản như **`EPS` (r=0.60)**, **`ROE` (r=0.61)** và tâm lý thị trường (sự kỳ vọng) **`P/E` (r=0.82)**.
  - Một phát hiện quan trọng là thị trường **"thưởng" cho sự an toàn**: Giá tương quan nghịch mạnh với **`Nợ/Vốn chủ sở hữu` (r=-0.60)** và tương quan thuận với **`Khả năng thanh toán` (r=0.59)**. Điều này có nghĩa là, khi FPT càng trở nên an toàn (giảm nợ, tăng khả năng trả nợ), giá cổ phiếu càng có xu hướng tăng.

## 3. Về Dữ liệu được sử dụng

# III - CHUẨN BỊ DỮ LIỆU


### 1 - Tiến hành xử lý những giá trị bị thiếu trong hai bảng dữ liệu


Các giá trị `NaN` được xử lý bằng cách điền giá trị `0` (sử dụng `fillna(0)`).

Việc này được thực hiện vì các giá trị thiếu này chỉ xuất hiện ở dòng dữ liệu cũ nhất, nơi không có dữ liệu của ngày hôm trước để so sánh. Gán giá trị `0` giúp bộ dữ liệu được nhất quán mà không ảnh hưởng đến việc phân tích.


#### 1.1 - Xử lý giá bị bị thiếu ở bảng data_price_history_FPT


```python
data_price_history_FPT["THAY ĐỔI GIÁ"] = data_price_history_FPT["GIÁ ĐÓNG CỬA"] - data_price_history_FPT["GIÁ ĐÓNG CỬA"].shift(-1)
data_price_history_FPT["% THAY ĐỔI"] = (data_price_history_FPT["THAY ĐỔI GIÁ"] / data_price_history_FPT["GIÁ ĐÓNG CỬA"].shift(-1)) 
data_price_history_FPT[["THAY ĐỔI GIÁ", "% THAY ĐỔI"]] = data_price_history_FPT[["THAY ĐỔI GIÁ", "% THAY ĐỔI"]].fillna(0)
```

#### 1.3 - Chia nhỏ dữ liệu data_financial_indicator_FPT thành các file dữ liệu để dễ phân tích


- Bảng dữ liệu `data_financial_indicator_FPT` lúc này là một bảng lớn chứa 8 nhóm chỉ số khác nhau.
- **Thực hiện các bước để chia nhỏ dữ liệu**
  1.  Xác định 8 dòng tiêu đề chuẩn (ví dụ: "Báo cáo kết quả kinh doanh", "Bảng cân đối",...).
  2.  "Quét" qua cột `CHỈ TIÊU` để tìm vị trí (số thứ tự dòng) của các tiêu đề này.
  3.  Dùng các vị trí đã tìm được để "cắt" bảng dữ liệu lớn thành 8 `DataFrame` (bảng) nhỏ riêng biệt.
- **Kết quả:** 8 bảng dữ liệu mới này được lưu thành 8 tệp Excel riêng biệt trong thư mục `/data/output/` và đồng thời được gán vào 8 biến mới (như `df_income`, `df_balance`,...) để có thể sử dụng ngay lập tức cho các phân tích sau.


```python
expected_sections = [
    "Báo cáo kết quả kinh doanh (Tỷ đồng)",
    "Bảng cân đối (Tỷ đồng)",
    "Chỉ số định giá",
    "Chỉ số hiệu quả hoạt động",
    "Chỉ số hiệu suất hoạt động",
    "Chỉ số cơ cấu nguồn vốn",
    "Chỉ số khả năng thanh toán",
    "Thông tin doanh nghiệp"
]

col_name = "CHỈ TIÊU"
if col_name not in data_financial_indicator_FPT.columns:
    raise KeyError(f"Không tìm thấy cột '{col_name}' trong file — cột hiện có: {list(data_financial_indicator_FPT.columns)[:10]}")

def normalize_text(s):
    if pd.isna(s):
        return ""
    return re.sub(r"\s+", " ", str(s)).strip().lower()

norm_col = data_financial_indicator_FPT[col_name].fillna("").astype(str).map(normalize_text)
patterns = []

for s in expected_sections:
    core = re.sub(r"chỉ tiêu", "", s, flags=re.I).strip()
    core_norm = normalize_text(core)
    patterns.append(core_norm)

section_indices = []
section_found_names = []
for i, txt in enumerate(norm_col):
    for j, pat in enumerate(patterns):
        if pat and pat in txt:
            section_indices.append(i)
            section_found_names.append(expected_sections[j])
            break

if len(section_indices) < len(expected_sections):
    fallback_idx = [i for i, txt in enumerate(norm_col) if any(k in txt for k in ["chỉ tiêu", "chỉ số", "thông tin doanh nghiệp"])]
    merged = sorted(set(section_indices + fallback_idx))
    section_indices = merged

section_indices = sorted(section_indices)
section_indices.append(len(data_financial_indicator_FPT))

n_found = len(section_indices) - 1

if n_found != len(expected_sections):
    print("CẢNH BÁO: Số đoạn tìm được khác với 8 (expected).")
    print("Các dòng tiêu đề phát hiện (index -> giá trị gốc):")
    for idx in sorted(set(section_indices[:-1])):
        print(idx, "->", data_financial_indicator_FPT.loc[idx, col_name])

sections = {}
for i in range(len(section_indices) - 1):
    start = section_indices[i] + 1 
    end = section_indices[i + 1]

    df_part = data_financial_indicator_FPT.iloc[start:end].reset_index(drop=True)
    name = expected_sections[i] if i < len(expected_sections) else f"Section_{i+1}"
    sections[name] = df_part

save_to_files = True
out_prefix = "./data/output/FPT_financial_"
if save_to_files:
    import os
    os.makedirs(os.path.dirname(out_prefix), exist_ok=True)
    for name, df in sections.items():
        safe_name = re.sub(r"[^\w\s-]", "", name).strip().replace(" ", "_")
        outpath = f"{out_prefix}{safe_name}.xlsx"
        df.to_excel(outpath, index=False)
        print("Saved:", outpath)

df_income = sections.get("Báo cáo kết quả kinh doanh (Tỷ đồng)")
df_balance = sections.get("Bảng cân đối (Tỷ đồng)")
df_valuation = sections.get("Chỉ số định giá")
df_efficiency = sections.get("Chỉ số hiệu quả hoạt động")
df_productivity = sections.get("Chỉ số hiệu suất hoạt động")
df_capstructure = sections.get("Chỉ số cơ cấu nguồn vốn")
df_liquidity = sections.get("Chỉ số khả năng thanh toán")
df_company_info = sections.get("Thông tin doanh nghiệp")
```

```python
print("Dữ liệu bảng Báo cáo kết quả kinh doanh:")
display(df_income)  

print("Dữ liệu bảng Bảng cân đối:")
display(df_balance)

print("Dữ liệu bảng Chỉ số định giá:")
display(df_valuation)

print("Dữ liệu bảng Chỉ số hiệu quả hoạt động:")
display(df_efficiency)

print("Dữ liệu bảng Chỉ số hiệu suất hoạt động:")
display(df_productivity)

print("Dữ liệu bảng Chỉ số cơ cấu nguồn vốn:")
display(df_capstructure)

print("Dữ liệu bảng Chỉ số khả năng thanh toán:")
display(df_liquidity)

print("Dữ liệu bảng Thông tin doanh nghiệp:")
display(df_company_info)
```

# IV - PHÂN TÍCH DỮ LIỆU CÔNG TY FPT


## 1 - Quan sát để hiểu doanh nghiệp và dữ liệu


### 1.1 Thống kê miêu tả

Thực hiện thống kê miêu tả cho hai bảng dữ liệu, bao gồm:

- **Five-number summary:** giá trị nhỏ nhất (min), tứ phân vị thứ nhất (Q1), trung vị (Q2/median), tứ phân vị thứ ba (Q3), giá trị lớn nhất (max).

#### 1.1.11 - Tạo một bảng dữ liệu mới từ hai bảng data_financial_indicator_FPT và data_price_history_FPT


```python
data_FPT_price_history = pd.read_excel('./data/Simplize_FPT_PriceHistory_20250315.xlsx', skiprows=5, nrows=1001)
data_FPT_indicator = pd.read_excel('./data/Simplize_FPT_FinancialIndicator_20250315.xlsx', skiprows=6, nrows=63)

data_FPT_price_history["NGÀY"] = pd.to_datetime(data_FPT_price_history["NGÀY"], errors="coerce")

def covert_period(col_name: str) -> pd.Period:
    text = col_name.strip().upper()
    match = re.search(r'Q([1-4])[\s/]*([0-9]{4})', text)
    if not match:
        match = re.search(r'QUÝ\s*([1-4])[\s/]*([0-9]{4})', text)
    if match:
        q, y = match.groups()
        return pd.Period(f"{y}Q{q}", freq="Q")
    else:
        raise ValueError(f"Không thể chuyển đổi kỳ '{col_name}' sang Period")

def convert_quarter_year(period: pd.Period) -> str:
    return f"Q{period.quarter} {period.year}"

new_cols = {}
for c in data_FPT_indicator.columns:
    if c != "CHỈ TIÊU":
        try:
            p = covert_period(c)
            new_cols[c] = convert_quarter_year(p)
        except Exception:
            pass
data_FPT_indicator.rename(columns=new_cols, inplace=True)

set_Q_period = []
for col in data_FPT_indicator.columns:
    if col != "CHỈ TIÊU":
        period = covert_period(col)
        if period.year >= 2021:
            set_Q_period.append(period)
set_Q_period = sorted(set_Q_period)

data_FPT = {convert_quarter_year(q): [] for q in set_Q_period}

important_features_of_price_history = ["GIÁ ĐÓNG CỬA", "THAY ĐỔI GIÁ", "% THAY ĐỔI"]
important_features_of_financial_indicator = [
    "Biên lợi nhuận gộp", "Biên lợi nhuận ròng", "P/E", "EPS", "Tăng trưởng EPS",
    "ROE", "Nợ phải trả / Vốn chủ sở hữu", "Khả năng thanh toán tổng quát",
    "Vòng quay tài sản", "Giá trị sổ sách"
]

avg_close_by_quarter = {}

for Q_period in set_Q_period:
    Q_last_day = Q_period.to_timestamp(how="end").date()
    start_day = Q_last_day - pd.Timedelta(days=14)
    end_day = Q_last_day + pd.Timedelta(days=14)
    start_time, end_time = pd.Timestamp(start_day), pd.Timestamp(end_day)

    filtered_data = data_FPT_price_history[
        (data_FPT_price_history["NGÀY"] >= start_time) &
        (data_FPT_price_history["NGÀY"] <= end_time)
    ]
    
    if filtered_data.empty:
        avg_close_by_quarter[convert_quarter_year(Q_period)] = np.nan
    else:
        avg_close_by_quarter[convert_quarter_year(Q_period)] = filtered_data["GIÁ ĐÓNG CỬA"].mean()

price_changes = {}
prev_close = None

for Q in set_Q_period:
    q_str = convert_quarter_year(Q)
    curr_close = avg_close_by_quarter[q_str]

    if prev_close is None or pd.isna(prev_close) or pd.isna(curr_close):
        change = np.nan
        pct_change = np.nan
    else:
        change = curr_close - prev_close
        pct_change = (change / prev_close) * 100 if prev_close != 0 else np.nan

    price_changes[q_str] = {"THAY ĐỔI GIÁ": change, "% THAY ĐỔI": pct_change}
    prev_close = curr_close

for Q_period in set_Q_period:
    q_str = convert_quarter_year(Q_period)
    
    avg_close = avg_close_by_quarter[q_str]
    change = price_changes[q_str]["THAY ĐỔI GIÁ"]
    pct_change = price_changes[q_str]["% THAY ĐỔI"]

    price_features = [avg_close, change, pct_change]

    fin_features = []
    for feature in important_features_of_financial_indicator:
        Feature_frame = data_FPT_indicator.loc[
            data_FPT_indicator["CHỈ TIÊU"].str.contains(feature, case=False, regex=True)
        ]
        if not Feature_frame.empty and q_str in Feature_frame.columns:
            fin_features.append(Feature_frame[q_str].values[0])
        else:
            fin_features.append(np.nan)
    
    data_FPT[q_str] = price_features + fin_features

New_Data_FPT = pd.DataFrame(data_FPT).T
New_Data_FPT.columns = important_features_of_price_history + important_features_of_financial_indicator

display(New_Data_FPT)
```

Đây là bộ dữ liệu **quan trọng nhất** của dự án, được tạo ra bằng cách kết hợp và chuẩn hóa cả hai bộ dữ liệu gốc. Bảng này đã:

Bộ dữ liệu mới này đã được xây dựng thành công. Nó cho thấy rõ nét mối liên hệ giữa giá thị trường (đã được làm mịn theo quý) và các chỉ số sức khỏe tài chính nội tại. Mọi thứ đã sẵn sàng cho bước trực quan hóa và tính toán ma trận tương quan.

        - **Ý nghĩa:** Gần như không có tương quan. Điều này cho thấy thị trường không phản ứng với _tốc độ tăng trưởng_ của EPS, mà phản ứng với **giá trị tuyệt đối** của EPS (tương quan 0.60). Miễn là FPT vẫn tạo ra EPS cao và ổn định, thị trường sẽ hài lòng.


### KẾT LUẬN

Dựa trên toàn bộ phân tích, **giá cổ phiếu FPT có mối quan hệ phụ thuộc chặt chẽ vào các yếu tố sau:**

1.  **Khả năng sinh lời tuyệt đối:** Giá FPT tăng khi **`EPS`** và **`ROE`** (hiệu quả sinh lời) tăng.
2.  **Hiệu quả hoạt động:** Giá FPT tăng khi công ty **`Vòng quay tài sản`** nhanh hơn (hoạt động hiệu quả hơn).
3.  **Sức khỏe tài chính (Giảm rủi ro):** Giá FPT tăng khi **`Khả năng thanh toán`** tăng và **`Tỷ lệ Nợ/Vốn chủ sở hữu`** giảm.
4.  **Tâm lý thị trường:** Giá FPT tăng khi thị trường sẵn sàng trả **`P/E`** cao hơn, phản ánh kỳ vọng vào tương lai.

**Ngược lại, giá cổ phiếu FPT _không_ phụ thuộc vào:**

- **Biên lợi nhuận ròng:** Vì chỉ số này của FPT rất ổn định và gần như không đổi.
